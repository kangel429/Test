```python
%matplotlib inline
from PIL import Image
import skimage
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import imageio
```

```python
im = Image.open('1.jpg').convert("LA")
```

```python
im.save("gray.png")
```

```python
num =0;
```

```python
row = im.size[0]
```

```python
col = im.size[1]
```

```python
gamma = 0.5
```

```python
result_img1 = Image.new("L", (row,col))
```

```python
while num < 11:
   
    for x in range(1,row):
        for y in range(1,col):
                value = pow(im.getpixel((x,y))[0]/255,(1/gamma))*255
                if value >=255:
                    value = 255
                result_img1.putpixel((x,y), int(value))
    result_img1.save("gamma_"+str(round(gamma,1))+".png")  # 감마  이미지 저장
    gamma = gamma+0.1
    num = num+1
```

```python
plt.imshow(result_img1,'gray')
```
