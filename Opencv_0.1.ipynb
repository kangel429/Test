```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import skimage
import imageio
import cv2 as cv
```

```python
url = 'https://upload.wikimedia.org/wikipedia/commons/0/08/Unequalized_Hawkes_Bay_NZ.jpg'
im = skimage.io.imread(url)
im.shape
```

```python
plt.imshow(im)
skimage.io.imsave('./original.png',im)
```

```python
im = im[:,:,0]
```

```python
plt.imshow(im, cmap='gray')
```

```python
count = np.zeros(256, dtype =np.int)
count.dtype
```

```python
for r in range (im.shape[0]):#row size
    for c in range(im.shape[1]):#col size
        pv = im[r,c]
        #print(pv)
        count[pv]+=1
```

```python
print(count)
```

```python
count.max(),count.argmax()
```

```python
count[137]
```

```python
plt.plot(count)
```

```python
def linearmap(x,xmin,xmax,tmin=0,tmax=255):
    t = ((x-xmin) / (xmax-xmin)) * (tmax-tmin) + tmin
    return t
```

```python
linearmap(137,100,210)
```

```python
i2 = np.empty_like(im)
i2.shape,im.shape
```

```python
for r in range(im.shape[0]):
    #print(r)
    for c in range(im.shape[1]):
        pv = im[r,c]
        tv = linearmap(pv,100,210)
        #print(pv,tv)
        i2[r,c] = tv
```

```python
plt.imshow(i2,cmap='gray')
i2.dtype
skimage.io.imsave('./something.png',i2)
```
