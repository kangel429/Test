```python
def linearmap(x, xmin, xmax,tmin=0, tmax =255):
    t = ((x-xmin)/(xmax-xmin)) * (tmax-tmin) + tmin;
    return t
    
```

```python
linearmap(100,150,230)
```

```python
i2 = np.empty_like(im)
i2.shape, im.shape
```

```python
for r in range(im.shape[0]):
    for c in range(im.shape[1]):
        pv = im[r,c]
        tv = linearmap(pv,100,210)
        i2[r,c] = tv
```

```python
plt.imshow(i2,cmap ='gray')
i2.dtype
skimage.io.imsave('./1.png',i2)
```
